```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv('product.csv')
df = df.iloc[:, 1:]
df
```

```python
sorted_df = df.sort_values(by=['Rating', 'Price'], ascending=[False, True])
sorted_df
```

```python
import pandas as pd
import numpy as np
import re

def extract_specifications(text):
    # Replace '•' with a comma for better splitting
    text = text.replace('•', ',')

    # Extract key-value pairs using regular expressions
    matches = re.findall(r'([^\s:,]+)\s*:\s*([^,]*)', text)
    
    # Create a dictionary from the extracted pairs
    specs = {key.strip(): value.strip() if value else np.nan for key, value in matches}

    return specs


# Applying the cleaning and extraction function to the 'Specifications' column
df['Specifications'] = df['Specifications'].apply(extract_specifications)

# Print the first few rows to inspect the data
print(df.head())

# Creating new columns
df['Battery'] = df['Specifications'].apply(lambda x: x.get('Battery', np.nan))
df['RAM'] = df['Specifications'].apply(lambda x: x.get('RAM', np.nan))
df['Camera'] = df['Specifications'].apply(lambda x: x.get('Camera', np.nan))
df['Display'] = df['Specifications'].apply(lambda x: x.get('Display', np.nan))

# Adding more specifications as needed

# Creating a 'PTA_Approved' column
df['PTA_Approved'] = df['Name'].str.contains('PTA Approved', case=False)

# Dropping the 'Specifications' column as it's no longer needed
df.drop('Specifications', axis=1, inplace=True)

df
```

```python
db_file = 'products.db'

# Connect to the SQLite database
conn = sqlite3.connect(db_file)

# Create a table and insert the DataFrame into it
df.to_sql('products', conn, index=False, if_exists='replace')

# Commit changes and close the connection
conn.commit()
conn.close()

print(f"DataFrame has been successfully inserted into the 'products' table in the '{db_file}' database.")
```
